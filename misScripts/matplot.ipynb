## Personalización Avanzada de Gráficos en Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
products = ["Producto A","Producto B","Producto C"]
market_share = np.random.randint(20, 150, 3)

plt.figure(figsize=(10,8))
plt.pie(
  x=market_share,
  labels=products,
  autopct="%1.2f%%",
  startangle=140,
  explode=(0,0,0.1),
  shadow=True,
  colors=["gold", "lightcoral", "orange"]
)

plt.title("Participación de mercado por producto")
plt.axis("equal")
```

## Visualización de Series Temporales con Python: Matplotlib y Pandas


```python
##Importando modulos
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.dates import DateFormatter
```

```python
##Generación de datos
dates = pd.date_range(start='2023-01-01', periods = 100)
values = np.random.rand(100).cumsum()
data = pd.DataFrame({'Date': dates, 'Value': values})

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(data['Date'], data['Value'], color='green')

plt.title('Serie de tiempo con formato en las fechas.')
plt.xlabel('Date')
plt.ylabel('Value')

plt.show()
```

```python
##De neración de datos mensules mediante una serie de tiempo
dates = pd.date_range(start='2023-01-01', periods=12, freq='ME')
sales = np.random.randint(1000, 5000, size=12)

sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})

plt.plot(sales_data['Date'], sales_data['Sales'], marker='o',
    linestyle='--', label='Ventas mensuales')

plt.gca().xaxis.set_major_formatter(DateFormatter("%b %Y"))

plt.xticks(rotation=45)

plt.title('Analisis de ventas mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()

plt.show()
```

## Creación de Gráficos Combinados en Matplotlib con GridSpec


```python
##Exportar librerias
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec  as gridspec
```

```python
## generate data simulation
x = np.linspace(0,10,100)
y = np.sin(x)

data = np.random.randn(100)
```

```python
## Genera grid space
gs = gridspec.GridSpec(2, 2, height_ratios=[2, 1], width_ratios=[1, 1])
fig = plt.figure(figsize=(10,  8))

## Primer subplot grande, ocupe tada la primer fila.
ax1  = fig.add_subplot(gs[0, :])
ax1.plot(x, y, color='blue')
ax1.set_title('Seno de X')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

## Segundo subplot, ocupa la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1, 0 ])
ax2.hist(data, bins=20, color='purple', edgecolor='black')
ax2.set_title('Histograma')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuancia')

## Tercer subplot, ubicado en la esquina inferior
ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, color= 'red')
ax3.set_title('Histograma')
ax3.set_xlabel('Valor')
ax3.set_ylabel('Frecuancia')

## Mostasr Grafico
plt.tight_layout()
plt.show()

```
