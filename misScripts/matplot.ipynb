## MatploLib

```python
## Exportar bibliotecas
import numpy as np
import matplotlib.pyplot as plt

print('Todo listo...')
```

```python
month = np.array(['Ja', 'Fe', 'Mar', 'A', 'Ma'])
sales = np.array([20, 25, 30, 28, 35])

#Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

#Crear gráfico
plt.plot(month, sales, marker='o', color='blue')

plt.title('Ventas mensuales de un producto')
plt.xlabel('Meses')
plt.ylabel('Ventas en miles de unidades')

plt.show()
```

```python
#Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

hours = [i for i in range(2, 9) ]
exam = [i*5+50 for i in range(1,8)]

plt.scatter(hours, exam, color='green')

plt.title('Relación entre horas estudiadas y el puntuaje')
plt.xlabel('Horas')
plt.ylabel('Puntuaje')

plt.show()
```

## Personalización Avanzada de Gráficos en Matplotlib


```python
#Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

hours = [i for i in range(2, 9) ]
exam_student_1 = [i*5+50 for i in range(1,8)]
exam_student_2 = [i*4+46 for i in range(1,8)]

## Crear un gráfico de dispersión de dos estudiantes.
plt.scatter(hours, exam_student_1, marker='o', color='green', linestyle='-', label='Estudiante 1')
plt.scatter(hours, exam_student_2, marker='s', color='red', linestyle='-', label='Estudiante 2')

plt.title('Relación entre horas estudiadas y el puntuaje')
plt.xlabel('Horas')
plt.ylabel('Puntuaje')

plt.show()
```
