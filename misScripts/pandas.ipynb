# Transformaciones y Manipulación de Datos con Pandas en Python

```python
import pandas as pd
path = 'Online_Retail.csv'
df = pd.read_csv(path, encoding='ISO-8859-1')
## Cálculo de Precio Total.
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print(df.head())
```

```python
## Filtrar valores mayores a 16
df['HighValue'] = df['TotalPrice'] > 16
print(df['HighValue'].head(15))
```

```python
## Da información de las conlumnas 
print(df.info())
```

```python
## Transfoma el tipo de dato de una columna.
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df.info())
```

```python
##  Aplicar función lambda
df['DiscountedPrice'] = df['UnitPrice'].apply(lambda x: x*0.9)
print(df.head(3))
```

```python
## Clasificación de precios unitarios
def categorize_price(price): 
    if price >= 4:
        return 'High'
    elif 2 <= price < 4:
        return 'Medium'
    else:
        return 'Low'
    
df['PriceCategory'] = df['UnitPrice'].apply(categorize_price)

print( df.loc[20:40,['UnitPrice', 'PriceCategory'] ] )
```

# Análisis de Ventas con Pandas: Agrupaciones y Estadísticas

```python
## Groupby - agrupar paises y cuenta 
country_count= df['Country'].value_counts()
print(country_count)
```

```python
## Agrupa los paises y suma la cantidad en este caso las ventas.
country_group = df.groupby('Country')['Quantity'].sum()
print(country_group)
```

```python
## Media y suma del precio unitario
country_stats = df.groupby('Country')['UnitPrice'].agg(['mean', 'sum']) 
print(country_stats)
```

```python
## Stock de compras
country_stock_group = df.groupby(['Country', 'StockCode'])['Quantity'].sum()
print(country_stock_group)
```

```python
def total_revenue(group):
    return (group['Quantity'] * group['UnitPrice']).sum()

revenue_per_count = df.groupby('Country').apply(total_revenue)
print(revenue_per_count)

```

```python
#Top 3 de mejores paises en ventas
top_3_best = revenue_per_count.sort_values(ascending=False)
top_3_worst = revenue_per_count.sort_values(ascending=True)
print("Top 3 peores: \n", top_3_worst.head(3))
print("Top 3 mejores: \n", top_3_best.head(3))

```

## Filtrado de Datos en Pandas para Análisis Efectivo


```python
import pandas as pd

## Leer el dataset de ventas desde un archivoCSV
file_path = "./Online_Retail.csv"
sales_data = pd.read_csv(file_path, encoding='ISO-8859-1')

## Convertir la columna 'InvoiceDate'  a tipo datetime
sales_data['InvoiceDate'] = pd.to_datetime(sales_data['InvoiceDate'])

## Eliminar filas con valores faltantes en la columnas críticas
sales_data.dropna(subset=['CustomerID', 'InvoiceDate'], inplace=True)

## Crear una nueva columna 'TotalPrice'
sales_data['TotalPrice'] = sales_data['Quantity'] * sales_data['UnitPrice']

print(sales_data.head())
```

```python
## Filtrar ventas en el reino unido.
uk_sales = sales_data[sales_data['Country'] == 'United Kingdom']
print(uk_sales)
```

```python
##Filtrar ventas altas (mayores a 10)
high_quantity_sales = sales_data[sales_data['Quantity'] > 10 ]
print(high_quantity_sales)
```

```python
## Filtrado de ventas por valor y pais especifico.
uk_kingdom_quantity_sales = sales_data[ (sales_data['Country']=='United Kingdom')&
                                        (sales_data['Quantity'] > 20) ]

print(uk_kingdom_quantity_sales)
```

```python
## Filtrado de información por año 
sales_2011 = sales_data[sales_data['InvoiceDate'].dt.year == 2011]
print(sales_2011)
```

```python
sales_dec_2010 = sales_data[(sales_data['InvoiceDate'].dt.year == 2010) &
                            (sales_data['InvoiceDate'].dt.month == 12)]
print(sales_dec_2010)
```

## Fusión de DataFrames con Pandas: merge, concat y join


```python
import pandas as pd
# Crear DataFrames de ejemplo
df1 = pd.DataFrame({
    'key':['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value2': [ 4, 5, 6 ]
})

print(df1)
print(df2)
```

```python
inner_merged = pd.merge(df1, df2, on='key', how='inner')
print(inner_merged)
```

```python
outer_merged = pd.merge(df1, df2, on='key', how='outer')
print(outer_merged)
```

```python
left_merged = pd.merge(df1, df2, on='key', how='left')
print(left_merged)
right_merged = pd.merge(df1, df2, on='key', how='right')
print(right_merged)
```

```python
# Crear DataFrames de ejemplo
df3 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
})
    
df4 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5']
})
    
# Concatenar verticalmente
vertical_concat = pd.concat([df3, df4])
print("Concatenación Vertical:\n", vertical_concat)
```

```python
# Concatenar horizontalmente
horizontal_concat = pd.concat([df3, df4], axis=1)
print("Concatenación Horizontal:\n", horizontal_concat)
```

```python
# Crear DataFrames de ejemplo con índices
df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

df6 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])

# Realizar un join
joined = df5.join(df6, how='inner')
print("Join con Índice:\n", joined)
```
