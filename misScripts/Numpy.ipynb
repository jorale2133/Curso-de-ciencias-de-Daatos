## Operaciones con Arrays en NumPy: Broadcasting, Concatenación y Más


Broadcasting

```python
prices = np.array([100, 200, 300])
discount = np.array([0.9])
discount_prices = prices*discount
print(discount_prices) #Salida: [ 90. 180. 270.]
```

```python
prices = np.random.randint(100, 500, size=(3, 3))
discount = np.array([10, 20, 30])
discount_prices = prices + discount

print(prices)
print(discount_prices)
```

```python
array = np.array([1,2,3,4,5])
print(np.all(array > 0)) #Todos losnumeros son mayores a cero
print(np.all(array > 4)) #Todos los numeros son mayores a 4
```

```python
print(np.any(array > 4)) #Existe  algun numero mayor a 4
```

Concatenación

```python
array_a = np.array([1,2,3])
array_b = np.array([4,5,6])
concatenar = np.concatenate((array_a,array_b))
print(concatenar)
```

Stacking

```python
satecked_v = np.vstack((array_a, array_b))
print(satecked_v) #Horizontal
```

```python
satecked_h = np.hstack((array_a,array_a))
print(satecked_h) #Vertical
```

Split

```python
array_c = np.arange(1,10)
split_array = np.split(array_c,3)
print(array_c)
print(split_array)
```

## Identificación y conteo de respuestas

```python
survey_responses = np.array(["bueno", "excelente", "malo", "bueno", "malo", "excelente",
                             "bueno", "bueno", "malo", "excelente", "excelente", "excelente"])
print(np.unique(survey_responses))
```

```python
unique_elements, counts= np.unique(survey_responses, return_counts=True)
print(unique_elements) #Bueno, excelente, malo 
print(counts) #[4 5 3]
```

Vista de variables(View)

```python
array_x = np.arange(10)
view_y  = array_x[1:3] 
print(array_x)
print(view_y)

array_x[1:3] = [10,11]
print(array_x)
print(view_y)
```

Copia de variables(Copy)

```python
array_x = np.arange(10)
copy_x = array_x[[1,2]]

print(array_x)
print(copy_x)

array_x[1:3] = [10, 11]
print( array_x)
print( copy_x)
```

## Manipulación de Matrices y Arrays en NumPy: Transponer, Invertir y Aplanar

Trasposición

```python
matrix = np.array([[1,2,3],[1,2,3],[7,8,9]])

transpoted_matrix = matrix.T
print(matrix)
print()
```

Reshape

```python
array = np.arange(1, 13)
reshaped_array = array.reshape(3, 4)
print(array)
print(reshaped_array)
```

Invertir array

```python
reversed_array = array[::-1]
print(array)
print(reversed_array)
```

Aplanado(Flattened)

```python
matrix = np.array([[1,2,3],[1,2,3],[7,8,9]])
flattened_array = matrix.flatten()
print(matrix)
print(flattened_array) 
```
