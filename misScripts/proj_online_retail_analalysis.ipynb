Poyecto de analisis de online retail

```python
import pandas as pd

## Cargar dataset
file_path = 'online_retail.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

print(data.describe())
print('Todo listo...')
```

```python
#Explorar datos
data.isnull().sum()
```

```python
#Explorar datos
data.duplicated().sum()
```

```python
## Etapa 2 - Liempaza datos
# Obtener valores unicos
unique_values={col: data[col].unique() for col in data.columns}

for col, values in unique_values.items():
    print(f'Columna:{col}')
    print(f'NÃºmero de valores Ãºnicos: {len(values)}')
    print(f'Valores unicos: {values[:10]}')
    print('-'*50)
```

```python
## Etapa 3 - Limpieza de datos
# quitar duplicados
data_cleaned = data.drop_duplicates()
data_cleaned = data_cleaned.dropna(subset=['CustomerID'])
print(data_cleaned)
```

```python
data_cleaned['TotalAmount'] = data_cleaned['Quantity'] * data_cleaned['UnitPrice']
data_cleaned.head(5)
```
